```python
import numpy as np
import pandas as pd
```

```python
Total=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
Total
```

```python
Active=Total.iloc[:,[1,-1]]
TotalCases=Active.groupby(['Country/Region']).sum()
TotalCases
```

```python
Deceased=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
Deceased
```

```python
Death=Deceased.iloc[:,[1,-1]]
Deaths=Death.groupby(['Country/Region']).sum()
Deaths
```

```python
Alive=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
Alive

```

```python
Recover=Alive.iloc[:,[1,-1]]
Recovered=Recover.groupby(['Country/Region']).sum()
Recovered
```

```python
New_Data=TotalCases.merge(Deaths, left_on='Country/Region', right_on='Country/Region')
New_Data
```

```python
Data=New_Data.merge(Recovered, left_on='Country/Region', right_on='Country/Region')
Data=Data.reset_index()
Data

```

```python
AliveCases=Data.iloc[:,1]-Data.iloc[:,2]-Data.iloc[:,3]
AliveCases

```

```python
Data["Alive"]=AliveCases
Data

```

```python
TotalCases=Data.iloc[:,1].sum()
Recovered=Data.iloc[:,2].sum()
Deaths=Data.iloc[:,3].sum()
Alive=Data.iloc[:,4].sum()
TotalCases
Recovered
Deaths
Alive
```
